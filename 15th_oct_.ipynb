```python
import pandas as pd
titanic_data={
    "passangerId":[1,2,3,4,5],
    "name":["john smith","jane doe","william brown","emily davis","micheal johnson"],
    'Age':[22,38,26,35,27],
    "sex":["male","female","male","female","male"],
    "pclass":[1,2,3,4,5],"survived":[0,1,1,0,0],
    "fare":[7.25,71.2833,7.925,53.1,8.05]
}
df=pd.DataFrame(titanic_data)
print(df.head())
```

```python
import pandas as pd
#display unique values before encoding
print(df['sex'].unique())
```

```python
df['sex']=df['sex'].map({'male':0,'female':1})
print(df['sex'])
```

```python
df_encode = pd.get_dummies(df, columns=['pclass'])
print(df_encode.head())

```
